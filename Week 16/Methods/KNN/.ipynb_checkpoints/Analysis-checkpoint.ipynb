```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
%run -i "../utils.py"
```

```python
all_measurements = []

for n in [1, 2, 3, 4]:
    print(f"N = {n} and P = 1")
    df_baseline_1_np1 = pd.read_csv(f"./ScoredBaseline-1-N{n}P1.csv")
    stats = df_baseline_1_np1.describe()
    print(stats)
    print()
    
    for stat in ['mean', 'min', '25%', '50%', '75%', 'max']:
        m = measure_method_by_threshold(
            output_fp=f"./ScoredBaseline-1-N{n}P1.csv",
            output_threshold=stats['Score'][stat],
            reverse=True,
            verbose=False
        )
        all_measurements.append((f"[N={n}, P=1]", stats['Score'][stat], m))

all_measurements.sort(key=lambda m: m[-1]['accuracy'], reverse=True)
for m in all_measurements:
    print(f"{m[0]} (Threshold: {m[1]}) Accuracy: {m[-1]['accuracy']* 100:.2f}%")
    cm = ConfusionMatrixDisplay(confusion_matrix=m[-1]['confusion_matrix'])
    cm.plot()
    plt.show()
```
