```python
import requests
import pandas as pd
import math
```

```python
df = pd.read_csv("../2/TEST_2/on_topic_sample_DeBERTa-v3-base-mnli-fever-anli.csv")
```

```python
ids = df.loc[df['UID'].str.len() >= 1]['UID'].tolist()
seen = [*ids]
ids = [(id_, 0) for id_ in ids]

MAX_HOPS = 5
MAX_NUM_CITATIONS = math.inf

while ids:
    curr_id = ids[-1][0]
    curr_hops = ids[-1][1]

    ids.pop()
    
    if curr_hops >= MAX_HOPS:
        continue
    
    out = requests.get(f"https://api.semanticscholar.org/graph/v1/paper/{curr_id}/citations").json()

    # No Citations
    if not out or not out.get('data', []):
        continue

    # Add Citations
    for i, citation in enumerate(out['data']):
        if i >= MAX_NUM_CITATIONS:
            break
        
        citation_id = citation['citingPaper']['paperId']
        
        # Don't Re-Visit
        if citation_id in seen:
            continue
        
        ids.append((citation_id, curr_hops + 1))
        seen.append(citation_id)

print(f"Number of Papers (Unfiltered): {len(seen)}")
```

```python
id_prefix = "https://openalex.org/"
id_prefix_len = len(id_prefix)

def extract_work_oa_id(work_url):
    return work_url[id_prefix_len:]

# ids = df.loc[df['LID'] != df['UID']]['LID'].tolist()
# seen = [extract_work_oa_id(id_) for id_ in ids]
# ids = [(extract_work_oa_id(id_), 0) for id_ in ids]

# print(ids)
# print(seen)

ids = [("W1972708686", 0)]
seen = ["W1972708686"]
```

```python
MAX_HOPS = 2
MAX_NUM_CITATIONS = math.inf

while ids:
    curr_id = ids[-1][0]
    curr_hops = ids[-1][1]

    ids.pop()

    print(("\t"*curr_hops) + curr_id)
    
    if curr_hops >= MAX_HOPS:
        continue
    
    out = requests.get(f"https://api.openalex.org/works/{curr_id}?select=referenced_works").json()

    # No Citations
    if not out or not out.get('referenced_works', []):
        continue

    # print(f"{curr_id} has {len(out['referenced_works'])} citations")
    
    
    # Add Citations
    for i, citation_id in enumerate(out['referenced_works']):
        if i >= MAX_NUM_CITATIONS:
            break

        # Looks like they're returned as URLs and not
        # the literal IDs.
        citation_id = extract_work_oa_id(citation_id)
        
        # Don't Re-Visit
        if citation_id in seen:
            continue
        
        ids.append((citation_id, curr_hops + 1))
        seen.append(citation_id)

print(f"Number of Papers (Unfiltered): {len(seen)}")
```
