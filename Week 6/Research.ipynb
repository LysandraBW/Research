```python
def get_vocab(text):
  doc = nlp(text)
  vocab = [t.lemma_ for t in doc if not t.is_stop and not t.is_punct and not ]
  return vocab
```

```python
def get_word_freq(text, graph=True):
  doc = nlp(text)
  vocab = [t.lemma_ for t in doc if not t.is_stop and not t.is_punct]

  word_freq = {}
  for word in vocab:
    if word not in word_freq:
      word_freq[word] = 0
    word_freq[word] += 1

  words = np.array(list(word_freq.keys()))
  counts = np.array(list(word_freq.values()))
  sorted = np.argsort(counts)

  if graph:
    plt.figure(figsize=(20, 20))
    plt.barh(words[sorted], counts[sorted])
    plt.title('Word Frequency')
    plt.xlabel('Word')
    plt.ylabel('Appearances')
    plt.show()

  return word_freq
```

```python
def compare(word_freqs):
  full_vocab = []
  for word_freq in word_freqs:
    full_vocab += list(word_freq.keys())

  cross_vocab = []
  for word in full_vocab:
    cross = True
    # Must Exist in All Vocabs
    for word_freq in word_freqs:
      if word not in word_freq:
        cross = False
        break
    if cross and word not in cross_vocab:
       cross_vocab.append(word)

  return cross_vocab
```
